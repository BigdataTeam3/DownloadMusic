```python
from bs4 import BeautifulSoup as bs
from decimal import *
path = 'C:/Users/BigData/Desktop/drum_template.mscx'
time = {
'measure':Decimal(1),
'whole':Decimal(1),
'half':Decimal(1)/2,
'quarter':Decimal(1)/4,
'eighth':Decimal(1)/8,
'16th':Decimal(1)/16,
'32nd':Decimal(1)/32,
'64th':Decimal(1)/64,
'128th':Decimal(1)/128
}
with open(path,'r') as f:
    music = bs(f.read(),'xml')
for measure in music.select('Score > Staff')[0].select('Measure'):
    print 'measure : '+measure.get('number')
    for notation in measure.find_all(['Rest','Chord']):
        dutime = notation.durationType.text
        print time[dutime]
        if notation.name == 'Rest':
            print '0'
        else:
#             print [n.text for n in notation.select('Note')]
            if '35' in [n.pitch.text for n in notation.select('Note')] or '36' in [n.pitch.text for n in notation.select('Note')]:
                print 'Drum'
```

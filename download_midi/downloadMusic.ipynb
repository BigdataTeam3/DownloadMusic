```python
# 國語歌

import requests
from bs4 import BeautifulSoup as bs
import pyodbc
import os
import time
url = 'http://sql.jaes.ntpc.edu.tw/javaroom/midi/alas/Ch/ch.htm'
# addr = 'C:/Users/BigData/Desktop/datas/java/Jfague/download/musictodb/國語1'
# fs = os.listdir(addr.decode('utf8').encode('big5'))
q = r'insert into musictable values (?,?,?,?)'
res = requests.get(url)
res.encoding = 'big5-hkscs'
soup = bs(res.text)
for a in soup.select('a'):
    if a.get('href') is None:
        continue
    url1 = '/'.join(url.split('/')[:-1])+'/'+a.get('href')
    res1 = requests.get(url1)
    res1.encoding = 'big5-hkscs'
    soup1 = bs(res1.text)
    for href in soup1.select('a'):
    #     print f.decode('big5').encode('utf8')
    #     filead = addr+'/'+f.decode('big5').encode('utf8')
        if len(href.text)==0:
            continue
    #     q = 'insert into musictable(mname, data, lang, src) values (?,?,?,?)'
        if href.get('href') is None:
            continue
        soundurl = '/'.join(url.split('/')[:-1])+'/'+href.get('href')
        res2 = requests.get(soundurl)
        res2.encoding='big5-hkscs'
        soup2 = bs(res2.text)
        for href2 in soup2.select('a'):
            if len(href2.text)==0 or href2.get('href') is None or href2.get('href').endswith('mid')==False:
                continue
            print href2.get('href')
            musicurl = '/'.join(soundurl.split('/')[:-1])+'/'+href2.get('href')
            musicname = ' '.join(href.text.split()).split('.')[0]
            if len(musicname)<1:
                continue
            conn = pyodbc.connect('DRIVER={SQL Server};SERVER=localhost;DATABASE=music;UID=sa;PWD=passw0rd')
    #         fi = open(filead.decode('utf8').encode('big5'), 'rb')
            content = requests.get(musicurl).content
            if len(content)<10:
                continue
    #         fi.close()
    #     conn = pymssql.connect(host='localhost:1433', user='sa', password='passw0rd', database='music', charset="utf8")
            cur = conn.cursor()
            insert = (musicname, pyodbc.Binary(content),unicode('國語'.decode('utf8')),unicode(url.decode('utf8')))
            cur.execute(q, insert)
            conn.commit()
            print a.text + ': ' + musicname + ' inserted'
            time.sleep(0.3)
cur.close()
conn.close()
```

```python
# input 資料型態
testkey = {u'staff1': {u'track1': '0.25,44,36;0.25,44,36;0.25,44;0.25,44,36', 'track0': '0.5,0;0.25,38;0.25,0'}}
```

```python
import Composer
from bs4 import BeautifulSoup as bs
newM = bs('')
staffid  = testkey.keys()[0].split('ff')[-1]  #取出 staff  id 數字
a = testkey[testkey.keys()[0]]
tkno =0
tkvalue = a.pop('track0')
newM.append(Composer.percussion(tkno,tkvalue)) 
if len(a) > 0:
    tick = newM.new_tag('tick')   # 建構Division tag
    tick.string= str(480*4*5)    # Division 480 * timesigN * measure number
    newM.append(tick)
for track in a.keys():
    tkno = int(track.split('k')[-1]) 
    tkvalue = testkey[testkey.keys()[0]][track]
    newM.append(Composer.percussion(tkno,tkvalue))
print newM.prettify()
```

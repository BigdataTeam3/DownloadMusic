# 寫音樂的原型,寫成mscx檔, 可延伸寫音樂特殊需求的檔案

```python
from bs4 import BeautifulSoup as bs  #open data

path = open('A Warm Place.mscx','r')
f = bs(path,'xml')
path.close()
```

```python
# xml: mean music file transform by beautifulsoup 全休止符 

def Rest(xml):
    Sig = xml.select('Measure')[0]
    Dz = int(Sig.select('sigN')[0].text)
    Dn = int(Sig.select('sigD')[0].text)
    dur = bs("<duration/>")
    dur.duration['n']=Dn
    dur.duration['z']=Dz
    rest = dur.duration.wrap(dur.new_tag("Rest"))
    durT = dur.new_tag("durationType")
    durT.string = "measure"
    dur.duration.insert_after(durT)
    return rest



```

```python

NewM = bs("")  #寫音樂的主體(body)

for i in range(1,5):
    NewM.append(NewM.new_tag("Measure",number = i))
    measure = NewM.select("Measure")[i-1]
    Ncho = NewM.new_tag('Chord')
    measure.append(Ncho)
    Ndur = NewM.new_tag('durationType')
    Ndur.string ="whole"
    measure.Chord.append(Ndur)
    Nnote = NewM.new_tag('Note')
    measure.Chord.append(Nnote)
    Ntie = NewM.new_tag('Tie',id=i)
    measure.Chord.Note.append(Ntie)
    Nend = NewM.new_tag('endSpanner',id=i)
    measure.Chord.Note.append(Nend)
    Npitch = NewM.new_tag('pitch')
    Npitch.string ="62"                # should double check "62"  could be run on xml file!!
    measure.Chord.Note.append(Npitch)
    Ntpc = NewM.new_tag('tpc')
    Ntpc.string="16"
    measure.Chord.Note.append(Ntpc)
    Nvel = NewM.new_tag('velocity')
    Nvel.string ="100"
    measure.Chord.Note.append(Nvel)
    veloT = NewM.new_tag('veloType')
    veloT.string ="user"
    measure.Chord.Note.append(veloT)
        
#     else: 
#         measure.append(Rest(f))
New =NewM.Measure.wrap(NewM.new_tag("Staff",id = 1)).prettify()
print New
# with open("trry.mscx", "wb") as file:
#     file.write(New)
# print "done"
```

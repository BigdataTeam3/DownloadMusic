## 產生音樂用

```python
import musicMaker
mc = musicMaker.MusicCreator()
mc.determineChords(['8g1','8g1','8g2','8g2'])
```

```python
# 打擊樂 Pattern
notations = ['A1','A1','A1','B1','A1','A1','A1','B1','A2','A2','A2','B2','A2','A2','A2','B2']
mli = mc.getMeasuresPerByNotations(notations,seed=7)
mc.addStaff(mli,'percussionType')
```

```python
# 吉他 Pattern
notations = ['A1','A1','A1','A1','A1','A1','A1','A1','A2','A2','A2','A2','A2','A2','A2','A2']
mgui = mc.getGuitarPitByNotations(notations)
mc.addStaff(mgui,instrument='guitar')
```

## Meta Information

```python
import xml.etree.ElementTree as et
from bs4 import BeautifulSoup as bs
def metaInfo(addr):
    fi = open(addr.decode('utf8').encode('big5'),'r')
    music = fi.read()
    fi.close()
    xml = bs(music,'xml')
    dic = {}
    division = xml.select('Division')[0]
    parts = xml.select('Part')
    
    for part in parts:
        for sube in part.select('Staff'):
            txt = ''
            txt += str(division)+'\n'
            txt += str(sube)+'\n'
            for adding in part.children:
                if adding.name != 'Staff':
                    txt += str(adding).strip()
            dic['Staff id='+sube.get('id')] = {'Meta':txt}
    return dic
```

## Measure Information

```python
from bs4 import BeautifulSoup as bs
def dataInfo(addr,metaDic):
    fi = open(addr.decode('utf8').encode('big5'),'r')
    music = fi.read()
    fi.close()
    xml = bs(music,'xml')
    dic = {}
    staffs = xml.select('Score > Staff')
    for staff in staffs:
        title = staff.name+' id='+staff.get('id')
        txt = ''
        m = ''
        measures = [m for m in staff.children if m.name is not None]
        for measure in measures:
            number = int(measure.get('number'))
            if number%4 != 0 and number != len(measures):
                m += str(number)+','
                txt += str(measure)+'\n'
            if number%4 == 0 or number == len(measures):
                m += str(number)
                txt += str(measure)+'\n'
                dic['Measure '+m] = txt
                m=''
                txt = ''
        metaDic[title].update(dic)
    return metaDic
```

```python
import xml.etree.ElementTree as et
from bs4 import BeautifulSoup as bs
addr = 'C:/Users/BigData/Desktop/All Out Of Love.mscx'
dic = metaInfo(addr)
result = dataInfo(addr,dic)
fi = open('C:/Users/BigData/Desktop/abc.txt','w')
fi.write(str(result))
fi.close()


```

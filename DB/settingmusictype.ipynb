```python
import pyodbc
import os
q = r'insert into musicname values (?,?)'
cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER=localhost;DATABASE=MUSICTABLE;UID=sa;PWD=passw0rd')
cur=cnxn.cursor()
for root, dirs, files in os.walk("e:/midieng"):
    for musicname in files:
        filepath='E:/midieng/'+musicname
        fi = open(filepath, 'rb')
        musicbinary = fi.read()
        print type(musicbinary)           
# print type("".join(map(bin,bytearray(musicbinary))).replace('b','x'))
# print "".join(map(bin,bytearray(musicbinary))).replace('b','x')
        insert=(musicname,pyodbc.Binary(musicbinary))
        cur.execute(q,insert)
        cnxn.commit() 
fi.close() 
cnxn.close()
```
